```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 50
DeltaT_values = [0.01, 0.05, 0.1]
colors = ['blue', 'green', 'orange']

# Function to simulate standard Brownian motion
def simulate_diffusion(sample_size, dt, total_time=5):
    steps = int(total_time / dt)
    trajectories = np.zeros((sample_size, steps))

    for j in range(sample_size):
        position = 0  # Initial position

        for i in range(1, steps):
            position = position + np.random.normal(0, np.sqrt(dt))
            trajectories[j, i] = position

    return trajectories

# Function to calculate MSD
def calculate_msd(trajectories, dt):
    N, steps = trajectories.shape
    displacement = trajectories[:, 1:] - trajectories[:, :-1]
    msd_values = np.mean(np.square(displacement), axis=0) * dt

    return msd_values

# Simulate and plot trajectories for each Δt in separate plots
plt.figure(figsize=(15, 5))

for i, dt in enumerate(DeltaT_values):
    trajectories = simulate_diffusion(sample_size, dt)
    msd_values = calculate_msd(trajectories, dt)
    plt.plot(np.arange(1, steps) * dt, msd_values, label=f'Dt={dt}', color=colors[i])

plt.title('Mean Squared Displacement (MSD)')
plt.xlabel('Time (seconds)')
plt.ylabel('MSD')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 10000  # No of trajectories
t = 5  # No of timesteps
dt_values = [0.01, 0.05, 0.1]
cases = ['deltaT = 0.01', 'deltaT = 0.05', 'deltaT = 0.1']
colors = ['blue', 'green', 'orange']

# Generate Trajectories and Calculate MSD
msd_values = []

for dt in dt_values:
    T = int(t / dt)
    r = np.sqrt(dt) * np.random.randn(T, N)
    x = np.cumsum(r, axis=0)
    msd = np.mean(x**2, axis=1)
    msd_values.append(msd)

# Plotting MSD for each case in separate subplots
fig, axes = plt.subplots(nrows=len(dt_values), figsize=(10, 20), sharex=True)

for i, (msd, case) in enumerate(zip(msd_values, cases)):
    axes[i].plot(np.arange(0, t, dt_values[i])[:len(msd)], msd, label=case, color=colors[i])
    axes[i].set_title(f'MSD for {case}')
    axes[i].set_xlabel('Time (seconds)')
    axes[i].set_ylabel('MSD')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
